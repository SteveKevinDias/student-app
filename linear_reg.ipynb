```python
# we have to change something in string we have to convert it in numerical
# we try to convert it into 0/1 with encoding technique
#Lets use label encoding
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
df["Extracurricular Activities"]
```

```python
# first create object of label encoder and then we call fit_transform along with column

le = LabelEncoder()
le.fit_transform(df["Extracurricular Activities"])
```

```python
#Update the dataframe
df["Extracurricular Activities"] = le.fit_transform(df["Extracurricular Activities"])
```

```python
df
```

```python
#Now we have create x and y
x = df[["Hours Studied",	"Previous Scores",	"Extracurricular Activities",	"Sleep Hours",	"Sample Question Papers Practiced"]]
```

```python
x
```

```python
y = df[["Performance Index"]]
```

```python
y
```

```python
#Split data into training and testing data
from sklearn.model_selection import train_test_split
```

```python
#itwill return Xtrain Ytrain Xtest Ytest
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train
```

```python
x_test
```

```python
#use training data to train the model
#Now there is some preprocessing to do before model building - 

scaler = StandardScaler()
x_trained_scaled = scaler.fit_transform(x_train)

```

```python
x_test_scaled = scaler.fit_transform(x_test)
```

```python
x_trained_scaled
```

```python
x_test_scaled
```

```python
# now we will do training

from sklearn.linear_model import LinearRegression
import pickle # physical file for deployment
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error # metrics
```

```python
# Now lets train model
model = LinearRegression()
model.fit(x_trained_scaled,y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
x_test_scaled
```

```python
model.predict([[-0.00230497, -1.23782039,  1.00702467,  0.29096856,  0.79816301]])
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
#Saving the model in the file
with open("linear_reg1_proj.pkl" , 'wb') as file:
    pickle.dump((model,scaler,le),file) # so that the model can understand the transformations and standardizations
```
